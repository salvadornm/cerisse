```python
import yt
import numpy as np
import matplotlib.pyplot as plt

def Plot1dBoxlib(files, ax):
  var_name = ["density", "xmom", "rho_E"]
  plt_title = [r"$\rho$", r"$\rho u$", r"$\rho E$"]

  for fname in files:
    ds = yt.load(fname)

    xaxis = 0  # take a line cut along the x axis
    lineout = ds.ortho_ray(xaxis, (0, 0))
    srt = np.argsort(lineout["index", "x"])
    x = np.array(lineout["index", "x"][srt])

    for i in range(3):
      ax[i].plot(x, np.array(lineout["boxlib", var_name[i]][srt]))
      ax[i].set_xlim([min(x), max(x)])
      ax[i].set_ylabel(plt_title[i])
      if i == 2:
        ax[i].set_xlabel("x")

  plt.tight_layout()
```

```python
fig, ax = plt.subplots(3, 1, figsize=[7,7])
axx = [ax[0], ax[1], ax[2]]
Plot1dBoxlib(["plt00000","plt01632","plt03283"], axx)
plt.legend([r"$t = 0.0$", r"$t = 1.0$", r"$t = 2.0$"])
plt.savefig("grid1024.png", facecolor='white')
```
